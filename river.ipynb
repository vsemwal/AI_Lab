```
#Importing Libraries
import tensorflow as tf
import numpy as np
import scipy as sp
```

```
datasetX = sp.ndimage.imread("river (1).png", flatten=True)

datasetY = np.zeros((64, 64))

for i in range(64):
    for j in range(64):
        if (datasetX[i][j] > 245):
            datasetY[i][j] = 1
        else:
            datasetY[i][j] = 0

datasetX = np.reshape(datasetX, [64 * 64, 1])
datasetY = np.reshape(datasetY, [64 * 64, 1])

alpha = 0.001


w0 = tf.Variable(np.random.randn(), dtype=tf.float32)

x0 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

modelLogit = tf.sigmoid(tf.multiply(x0, w0))

J = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(
    logits=modelLogit, labels=Y))

optimizer = tf.train.GradientDescentOptimizer(alpha).minimize(J)

epochs = 1000

with tf.Session() as session:
    session.run(tf.global_variables_initializer())

    for i in range(epochs):
        print("epoch: ", i)
        for (x, y) in zip(datasetX, datasetY):
            session.run(optimizer, feed_dict={x0: x, Y: y})

    print("Weights: ", session.run(w0))

    testX = sp.ndimage.imread("test64.png", flatten=True)

    testX = np.reshape(testX, [64 * 64, 1])

    outputY = []

    for i in range(64 * 64):
        predic = session.run(tf.sigmoid(tf.multiply(np.float32(testX[i]), w0)))

        if (predic >= 0.5):
            outputY.append(255)
        else:
            outputY.append(0)

    outputY = np.reshape(outputY, [64, 64])

    print("Output image saved.")
    sp.misc.imsave('../output64.png', outputY)
```
